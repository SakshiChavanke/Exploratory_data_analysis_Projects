```python
## Split Final Actor Column and adding them back to new dataframe

final_df_actor_expanded = df1['actors'].str.split('|', expand=True)
final_df_actor_expanded.columns = ['C'+str(i) for i in final_df_actor_expanded.columns]

## Concating old Final Dataframe & Expanded Actor dataframe
concated_final_df = pd.concat([df1, final_df_actor_expanded], axis = 1)

final_df1 = pd.melt(concated_final_df, id_vars1, value_vars=final_df_actor_expanded.columns, var_name='Actor Number', value_name='actors_split')
```

```python
## Split Final Genre Column and adding them back to new dataframe

final_df_genre_expanded = df1['genre'].str.split('|', expand=True)
final_df_genre_expanded.columns = ['C'+str(i) for i in final_df_genre_expanded.columns]

## Concating old Final Dataframe & Expanded Genre dataframe
concated_final_df = pd.concat([df1, final_df_genre_expanded], axis = 1)

final_df2 = pd.melt(concated_final_df, id_vars1, value_vars=final_df_genre_expanded.columns, var_name='Genre Number', value_name='genre_split')
```

```python
## Split Final Actor Column and adding them back to new dataframe

final_df_writer_expanded = df1['writers'].str.split('|', expand=True)
final_df_writer_expanded.columns = ['C'+str(i) for i in final_df_writer_expanded.columns]

## Concating old Final Dataframe & Expanded Actor dataframe
concated_final_df = pd.concat([df1, final_df_writer_expanded], axis = 1)

final_df3 = pd.melt(concated_final_df, id_vars1, value_vars=final_df_writer_expanded.columns, var_name='writer Number', value_name='writer_split')
```

Best Actor gives big hits

```python
actor_df = final_df1.sort_values(['hitFlop'],ascending=False).groupby('actors_split').head(5)
actor_df
```

```python
actor_df = final_df1[['actors_split', 'hitFlop']]
#actor_df = actor_df.groupby('actors_split').agg(['min', 'max'])
actor_df = actor_df.groupby("actors_split").agg(a_max=pd.NamedAgg(column="hitFlop", aggfunc="max"), b_min=pd.NamedAgg(column="hitFlop", aggfunc="min"),c_sum=pd.NamedAgg(column="hitFlop", aggfunc="sum"))
actor_df = actor_df.sort_values(['c_sum'],ascending=False)
actor_df
```

```python
plt.figure(figsize=(30,20))
actor_df.head(15).plot(kind='bar')
```

Each Director has his/her own genre

```python
bygenre= pd.DataFrame(final_df2.groupby('genre_split').directors.value_counts())
bygenre
```

```python
unstack_bygenre = bygenre.unstack().T
unstack_bygenre
```

```python
best_director_genrewise = unstack_bygenre.idxmax()
best_director_genrewise
```

```python
genre_director_score = unstack_bygenre.max()
genre_director_score
```

```python
# Best Action Director

df = pd.DataFrame(unstack_bygenre[unstack_bygenre.action == unstack_bygenre.action.max()])
action_df = df.T

plt.figure(figsize=(30,20))
action_df.plot(kind='bar')
```

```python
# Best Comedy Director

df = pd.DataFrame(unstack_bygenre[unstack_bygenre.comedy == unstack_bygenre.comedy.max()])
comedy_df = df.T

plt.figure(figsize=(30,20))
comedy_df.plot(kind='bar')
```

```python
# Best Drama Director

df = pd.DataFrame(unstack_bygenre[unstack_bygenre.drama == unstack_bygenre.drama.max()])
drama_df = df.T

plt.figure(figsize=(30,20))
drama_df.plot(kind='bar')
```

```python
# Best Romance Director

df = pd.DataFrame(unstack_bygenre[unstack_bygenre.romance == unstack_bygenre.romance.max()])
romance_df = df.T

plt.figure(figsize=(30,20))
romance_df.plot(kind='bar')
```
