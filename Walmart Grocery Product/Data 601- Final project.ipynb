# 6. Hypothesis Testing

### People consume More Alcohol which is Cheaper

```python
#Making Box plot to see the range of price for alcohols
sns.boxplot(data[ (data['CATEGORY'] == 'Wine') | (data['CATEGORY'] == 'Spirits') | (data['CATEGORY'] == 'Beer')]['PRICE_RETAIL'])
```

```python
#Finding maximum  value for defining range of beans 
data[ (data['CATEGORY'] == 'Wine') | (data['CATEGORY'] == 'Spirits') | (data['CATEGORY'] == 'Beer')]['PRICE_RETAIL'].describe().T
```

```python
#Defining different beans range
def make_bins(x):
    if x <= 10:
        return '0 to 10'
    if x <= 20:
        return '11 to 20'
    if x <= 30:
        return '21 to 30'
    if x <= 40:
        return '31 to 40'
    if x <= 50:
        return '41 to 50'
    if x <= 60:
        return '51 to 60'
    if x <= 70:
        return '61 to 70'
    if x <= 80:
        return '71 to 80'
    if x <= 90:
        return '81 to 90'
    if x <= 100:
        return '91 to 100'
    if x <= 110:
        return '101 to 110'
```

```python
#Adding data to a new dataframe for fetching alcoholic drinks
data['Price Bins'] = data['PRICE_RETAIL'].apply(lambda x: make_bins(x))

temp = data[ (data['CATEGORY'] == 'Wine') | (data['CATEGORY'] == 'Spirits') | (data['CATEGORY'] == 'Beer')]
temp['Price Bins'] = temp['PRICE_RETAIL'].apply(lambda x: make_bins(x))
```
